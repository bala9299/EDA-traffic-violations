# traffic violations maryland county
### 2012 - 2018

### summary


* 67% of ♂  creates traffic violations
* 25% of violation cases were represented by 4th district, Wheaton officers in court 
* zero accidents
* 3.4% of violations are involved by not wearing a seat belt and 2.2% of violations are occured by male drivers 
* 0.24% (3065) violations caused personal injury by not wearing seat belt 
* 2208 violations are commited by alcohol
* 94 violations caused personal injury by alcohol
* 1.9% of violations committed property damage
* 180 violations caused property damage by alcohol
* 279 violations involved fatality (zero accident need to inspect)
* there is no violation involved fatality by alcohol 
* 6622 violations involved by commercial vehicle 
* 4770 violations involved by commercial vehicle without licenses 
* 248 violations commited by work zone in road side
* 87.4% of violations involved by MA (maryland) registration vehicle 
* 87.5% of violations commited by automobile (cars)
* 21.9% of violations caused by toyota and honda vehicles
* 90% of traffic violations caused by maryland drivers
* 25% of violation commited by drivers of silverspring city , maryland state 
* 82% of violation cases arrested by marked patrol vehicle 
* 26% of ♀ arrested by marked patrol vehicle for traffic violations
* year 2015 has the most violations cases
* year 2018 has the lowest violations case 
* in the year of 2018 , there is no data after a month of may (reason why 2018 has the lowest cases)
* march and april month have filed a most number of violation cases in between the year of 2012 to 2018
* almost everyday the year 2015 filed most cases
* 47% of violations were warned by officers
* the most violations are commited by black and silver color vehicle 
* 5% of asian drivers commited traffic violations
